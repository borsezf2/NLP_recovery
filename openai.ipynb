````python
from openai import OpenAI
import time
import os
from ratelimit import limits, sleep_and_retry


# Set your OpenAI API key as an environment variable
os.environ["OPENAI_API_KEY"] = os.getenv('GIT_OPENAPI_KEY')
````

````python
# Define rate limit: 3 calls per minute
CALLS = 3
RATE_LIMIT = 60

def measure_time(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"{func.__name__} time: {end_time - start_time:.2f} seconds")
        return result
    return wrapper

@sleep_and_retry
@limits(calls=CALLS, period=RATE_LIMIT)
def generate_text(prompt, max_tokens=400, model="gpt-3.5-turbo"):
# def generate_text(prompt, max_tokens=400, model="gpt-4o"):
    client = OpenAI()
    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": "i will give you a problem related to a distributed application, resolve it and give a python code to resolve it so that no human input is needed."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=max_tokens,
    )
    return response.choices[0].message.content.strip()
````

````python

# Define your prompt

# Ask a coding question
prompt = "A host at ip 10.5.16.213 username harsh and password 123, has a unwanted process eating a lot of CPU usage"

# Example usage
# prompt = "What is the capital of France?"
try:
    for i in range(1):  # Try to make 5 calls
        result = generate_text(prompt)
        print(f"Call {i+1}: {result}")
except Exception as e:
    print(f"An error occurred: {e}")
````
